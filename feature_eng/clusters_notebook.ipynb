```python
import ast
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import pdist
```

```python
#load data
df_embed = pd.read_csv("../data_versions/embeddings.csv")

#turns str to py list
#df_embed["Embedding"] = df_embed["Embedding"].apply(ast.literal_eval) THIS DID NOT WORK FOR ME
def parse_numpy_array(s):
    try:
        return np.fromstring(s.strip('[]'), sep=' ')
    except Exception as e:
        print(f"Error parsing: {s}\n{e}")
        return None

df_embed['embedding'] = df_embed['Embedding'].apply(parse_numpy_array)

#turn to 2D np array
X = np.vstack(
    (
        df_embed['embedding']
    )
)
```

```python
X.shape
```

```python
#linkage matrix: Wards Method
Z = linkage(X, method = 'ward', metric = 'euclidean')
```

```python
#dendogram
plt.figure(figsize = (10,5))
dendrogram(Z)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Samples (Descriptions)")
plt.ylabel("Distance")
plt.show()
```

```python
df_embed.shape
```

```python
DIST_THRESHOLD = 6 #discuss with team
clusters = fcluster(Z, t = DIST_THRESHOLD, criterion = 'distance')
```

```python
np.unique(clusters)
```

```python
df_embed["Cluster"] = clusters
df_embed.head()
```

```python
df_to_save = df_embed.drop(['Embedding', 'embedding'], axis=1)
df_to_save.head()
```

```python
df_to_save.to_csv('../data_versions/clusters.csv')
```
